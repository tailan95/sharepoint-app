```python
# Table
dbutils.widgets.text("db_table", "None", "Tabela")
db_table = dbutils.widgets.get("db_table")
if not db_table or db_table.lower() == "none":
    raise ValueError("Tabela não fornecida ou inválida")

# Sharepoint link
dbutils.widgets.text("sharepoint_url", "None", "Sharepoint")
sharepoint_url = dbutils.widgets.get("sharepoint_url")
if not sharepoint_url or sharepoint_url.lower() == "none":
    raise ValueError("Nenhum link inserido")

# Mode
dbutils.widgets.text("mode", "overwrite", "Modo")
overwrite = dbutils.widgets.get("mode")
if overwrite not in ["overwrite", "append"]:
    raise ValueError("Nenhum link inserido")
else:
    overwrite = True if overwrite == "overwrite" else False
```

```python
import os
import sys
current_path = os.getcwd()
parent_path = os.path.dirname(current_path)
sys.path.append(parent_path)
```

```python
from core.SharepointIngest import SharepointIngest
sharepoint = SharepointIngest()
df = sharepoint.execute(file_url=sharepoint_url)
```

```python
from core.ProcessDF import ProcessDF
processing = ProcessDF()
df = processing.execute(df)
```

```python
df_spark = spark.createDataFrame(df)
df_spark.write \
    .format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite" if overwrite else "append") \
    .saveAsTable(db_table)
```
